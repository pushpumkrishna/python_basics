# Filling Missing Values ::

  * **Country Column** ::

```python
# Taking Country column first

print('Data type of the Country column is :: ', data['country'].dtypes)

# Since the type is object hence calculating the value count of column

print('Different values of the country column is:: \n', data['country'].value_counts())
```

```python
#creating a copy of original dataset

data_copy1 = data.copy()

# Dropping Reservation status date column as it won't help much

data_copy1.drop(['reservation_status_date'], axis = 1, inplace = True)
data_copy1
```

```python
# Trying the fillna method to see if are still missing values present


data_copy1['country'].fillna(method = 'ffill', limit = 1, inplace = True)    # limit: this is the max no of consecutive NaN values to ffill/bfill.
print("Total NaN values present in the column after using Forward-Fill method:: ",data_copy1['country'].isnull().sum())
```

* Since still 309 values are left to fill. Hence using Back Fill method once again

```python
data_copy1['country'].fillna(method = 'bfill', limit = 1, inplace = True)    # limit: this is the max no of consecutive NaN values to ffill/bfill.
print("Total NaN values present in the column after using Forward-Fill method:: ",data_copy1['country'].isnull().sum())
```

* Since still 219 values are left to fill. Hence using fillna method once again and filling it with mode value

```python
# Mode value of the country column 
mode(data_copy1['country']).mode[0]
```

```python
# Filling the missing values with the mode values

data_copy1['country'].fillna(value = mode(data_copy1['country']).mode[0], inplace = True)    
print("Total NaN values present in the column after using mode values:: ",data_copy1['country'].isnull().sum())
```
