```python
import pandas as pd

# loading the data set

data = pd.read_csv("E:Data Science/CSV dataset/StudentsPerformance.csv")
data
```

```python
# Doing F test to get f avalue and p value

import scipy.stats as stats

fvalue, pvalue = stats.f_oneway(data['math score'], data['reading score'], data['writing score'])
print(f"f value is: {fvalue} \np value is: {pvalue}")
```

```python
# arranging data in format suitable for ANOVA test using melt() function

data_melt = pd.melt(data.reset_index(), id_vars=['index'], value_vars = ['math score', 'reading score','writing score'])
data_melt.columns = ['index', 'treatments', 'value']
data_melt
    
```

```python
# Doing ANOVA Test

import statsmodels.api as sm
from statsmodels.formula.api import ols

model = ols('value ~ C(treatments)', data = data_melt).fit()
anova_table = sm.stats.anova_lm(model, typ = 2)
anova_table
```

```python
# TUkey HSD test to determine diff in various mean

from statsmodels.stats.multicomp import pairwise_tukeyhsd

tukey_hsd = pairwise_tukeyhsd(endog = data_melt['value'], groups = data_melt['treatments'], alpha = 0.05)
tukey_hsd
```
