```python
## Drawing the regplot 

import seaborn as sns

x = data['Gross capital']
y = data['GDP']
sns.regplot(x, y)
```

```python
## Calculating the f value and p value of the desired columns

import scipy.stats as stats

fval, pval = stats.f_oneway(data['Consumer public adm'], data['Gross capital'], data['Equip. Goods others'], data['Const.'])
print('f Value is:: %.5f \np Value is :: %.15f' %(fval, pval))
```

```python
## Converting the data into correct format

data_new = pd.melt(data.reset_index(), id_vars = 'index', value_vars = ['Consumer public adm', 'Gross capital', 'Equip. Goods others', 'Const.'])
data_new.columns = ['index', 'treatments', 'value']
data_new
```

```python
## Performing the anova test 

import statsmodels.api as sm
from statsmodels.formula.api import ols

model = ols('value ~ C(treatments)', data = data_new).fit()
anova_table = sm.stats.anova_lm(model, typ = 2)
print(anova_table)
```

```python
# Performing the Tukey HSD test to analyze the differences in the different means

from statsmodels.stats.multicomp import pairwise_tukeyhsd

hsd_test = pairwise_tukeyhsd(endog = data_new['value'], groups = data_new['treatments'], alpha = 00.05)
print(hsd_test)
```
