```python
# defining a function that returns and prints the total unique values and its count in that particular column as input

def count_values(x):
    """
    Input:: A column of the dataset passed as dataframe
    Output:: Total count of unique values in that column
    
    """
    values = x.value_counts()
    
    print('Total number of Unique values and its count in the column is:: \n\n', values)
    print()
```

```python
# Using count_values funtion to print the unique values in predictor variable

temp1 = data[['year']]
temp1.apply(count_values)
```

```python
# Using count_values funtion to print the unique values in title status  variable

temp2 = data[['title_status']]
temp2.apply(count_values)
```

```python
# Using count_values funtion to print the unique values in color variable

temp1 = data[['color']]
temp1.apply(count_values)
```
