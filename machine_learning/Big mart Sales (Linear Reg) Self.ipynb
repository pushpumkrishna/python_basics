```python
# Defining a function to calculate the total NaN values in the dataset/column

def nan_values(x):
    """
    Function to print all NaN values present in the dataset/column
    """
    
    val = x.isnull().sum()
    return val
    
```

```python
# Importing libraries to make the model

#for modelling
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

* Using two features, MRP and the store establishment year to estimate sales. Now, let us built a linear regression model in python considering only these two features.

```python
# Defining the predictor variable

X = train[['Item_MRP', 'Outlet_Establishment_Year']]

# Defining our target variable

Y = train['Item_Outlet_Sales']
```

```python
# dividing the data into training and validation data

x_train, x_val, y_train, y_val = train_test_split(X, Y, test_size = 0.2)

x_train.shape, x_val.shape, y_train.shape, y_val.shape
```

```python
# defining our model

model = LinearRegression()

# Fitting the data into the model and training the algorithm

model.fit(x_train, y_train)

# making the predictions on the validation data

y_predict = model.predict(x_val)

print('Predicted Values of the model are:: \n', y_predict)
```

```python
# Defining a function to calculate the Mean Squared Error

import numpy as np

def mse(x,y):
    """
    Function to return the calculated Mean Squared error of the model
    """
    
    error = round(np.mean((x-y)**2), 4)
    return error
    
```

```python
# Calling the mse function to calculate the MSE of the model

print("Mean Squared error of the model is:: ", mse(y_predict, y_val))
```

```python
# Calculating the coeficients of our equation

coeff = model.coef_

# Substituting the values in the dataframe

col = x_train.columns      # Defining the columns

coef_df = pd.DataFrame({'Input Variables' : col, 'Calculated Coefficients': coeff})
coef_df
```

# Observation:: 
  
  * We can see that MRP has a high coefficient, meaning items having higher prices have better sales.  

# R-Square: 

 * It determines how much of the total variation in Y (dependent variable) is explained by the variation in X (independent variable).
 * The value of R-square is always between 0 and 1, where 0 means that the model does not model explain any variability in the target variable (Y) and 1 meaning it explains full variability in the target variable.

```python
# Calculating the R square value

r_square = round(model.score(x_val, y_val), 4)
print('R-Square value of the model is:: ', r_square)
```

* RÂ² is 29%, meaning, only 29% of variance in sales is explained by year of establishment and MRP. 
* In other words, if you know year of establishment and the MRP, youâ€™ll have 29% information to make an accurate prediction about its sales.

```python
# Calling NaN missing function and checking to total NaN values present

temp1 = train[['Item_Weight']]
print('Total number of NaN values present in Item Weight column is:: ')
temp1.apply(nan_values)
```

```python
# Filling the missing values with the mean value of the weight column

weight_mean = round(np.mean(train['Item_Weight']), 4)
print('Average weight of the item weights is :: ', weight_mean)

# Substituting the mean values in the item weight column

train['Item_Weight'].fillna(weight_mean, inplace = True)
```
