```python
## Defining function to calculate values

list1 = [2,4,5,6,7,8,45,666,7,5,6,4,5433,5,6,545,6,76,45,6,545]


def min_and_max(data): 
    min_val = min(data)
    max_val = max(data)
    return (min_val, max_val)


def range_f(data):
    max_val = max(data)
    min_val = min(data)
    return(min_val, max_val)

print(min_and_max(list1))
print(range_f(list1))
```
