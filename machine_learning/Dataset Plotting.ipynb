```python
# Finding the correlation between the variables and plotting the heatmap to visualize it


x = data.corr().abs()                           ## taking only absolute values
fig, ay = plt.subplots(figsize = (15,10))
sns.heatmap(x, annot = True, ax =ay, cmap = 'coolwarm')
```
