```python
## Creating the dataframe 

import pandas as pd

columns = ['name', 'sex', 'age']

user1 = pd.DataFrame([['Kush', 'Male', '22'], ['Luv', 'Male', '21']], columns = columns)
print(user1)

user2 = pd.DataFrame([['Honey', 'Male', '25'], ['Logan', 'Male', '25']], columns = columns)
print('\n', user2)

user3 = user1 + user2    # Using this adds values beside each other
print(user3)
```

```python
user4 = pd.DataFrame(dict(name = ['Luv', 'Kush'],
                         age = ['11', '12'],
                         job = ['Doc', 'Engg.'],
                         sex = ['male', 'female']))
user4
```

```python
## Appending dataframes

user5 = user1.append(user2)
user5
```

```python
user5.append(user4)
```

```python
## All dataframes can be concatenated in single command

user6 = pd.concat([user1, user2, user4, user5])
user6
```

```python
## Merging two dataframes

output = pd.merge(user1, user2, on = 'name')
output
```
